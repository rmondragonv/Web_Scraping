# Punto 1:

- `[15 pts]` Hacer Web Scraping de 10 animales en Wikipedia (en búcle)
- `[10 pts]` Obtener el **encabezado** de cada animal
- `[15 pts]` Obtener todos los **textos** que están en las etiquetas de negrilla del primer párrafo.

```python
import bs4 as bs 
import urllib.request
```

```python
#Hacemos una lista para iterar con los links de los 10 animales
enlaces = ['https://es.wikipedia.org/wiki/Felis_silvestris_catus', 'https://es.wikipedia.org/wiki/Canis_familiaris', 'https://es.wikipedia.org/wiki/Hyaenidae', 'https://es.wikipedia.org/wiki/Tarentola_mauritanica', 'https://es.wikipedia.org/wiki/Le%C3%B3n', 'https://es.wikipedia.org/wiki/Panthera_tigris', 'https://es.wikipedia.org/wiki/Cebra', 'https://es.wikipedia.org/wiki/Giraffa_camelopardalis', 'https://es.wikipedia.org/wiki/Elefante', 'https://es.wikipedia.org/wiki/Hippopotamidae']
#Hacemos otra lista para agregar los encabezados 
encabezados =[]
#Hacemos otra lista para los palabras en negrilla
negrilla = []
```

```python
#Recorremos por id la lista enlaces
for i in range(len(enlaces)):
  #cargamos cada enlace en source
  source = urllib.request.urlopen(enlaces[i]).read()
  #Usamos soup como variable para el lxml
  soup = bs.BeautifulSoup(source,'lxml')
  #H1 siempre es el encabezado de la pagina así que tomamos H1
  tag = soup.h1
  #Lo agregamos a la lista con la cualidad string que es la que nos importa
  encabezados.append(tag.string)
  #Dado que bs4 con solo .p nos encuentra el primer parrafo no es nada más necesario
  primer_parrafo = soup.p
  #Para cada palabra en el primer parrafo buscamos todas las palabras que comiencen con la etiqueta de negrilla b
  for word in primer_parrafo.find_all('b'):
    negrilla.append(word.string)
#Imprimimos
for i in range(len(encabezados)):
  print("Encabezado: " + encabezados[i])
```

```python
for i in range(len(negrilla)):
  print("Palabras en negrilla: " + negrilla[i])
```

# Punto 2:
- `[10 pts]` Usando regex, reemplazar todas las caracteres especiales del punto anterior (palabras en negrilla) por un asterisco (¡Ojo, los espacios se quedan!)


```python
import re

for i in range(len(negrilla)):
  txt = negrilla[i]
  print(re.sub("[áéíóúñ]", "*" , txt))
```
